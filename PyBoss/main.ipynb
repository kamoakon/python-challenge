```python
import csv

import pandas as pd

import os
 
#cwd = os.getcwd()
#print(cwd)


csv_file = input('Enter the full path to your file ')


```

```python
df = pd.read_csv(csv_file, sep=',')
df.head()
```

```python
#Split first name and last name
df[['First Name', 'Last Name']] = df.Name.str.split(' ', expand = True)
df.head()

df.drop('Name', axis=1)

#Changing columns
df = df[["Emp ID","First Name","Last Name", "DOB", "SSN", "State"]]
```

```python
#Converting yyyy-mm-dd to dd/mm/yyyy format

df['DOB'] = pd.to_datetime(df['DOB'])

df['DOB'] = df['DOB'].dt.strftime('%m/%d/%Y')
df.head()
```

```python
#Working around the format to pull the last 4 digits and put the standard SSN format in front of the 4 numbers.

df['SSN'] = "***-**" + df['SSN'].str[-5:]

df.head()
```
